```python
# get number of rows with same region
df['region'].value_counts()
```

```python
#to portray the top 15
df['region'].value_counts()[:15]
```

```python
# make it a variable
top_15 = df['region'].value_counts()[:15]
```

```python
plt.figure(figsize=(12,8))
plt.title("Top 15 brands of vehicles", fontsize=14, fontweight='bold')
top_15.plot.bar()
plt.xlabel('Name of brand',fontsize=12, fontweight='bold')
plt.ylabel('Number of Vehicles',fontsize=12, fontweight='bold')
plt.grid()
# save the plot to file
fig = plt.gcf()
fig.savefig('top-15-brands.png')
plt.show()
```

```python
df['brand'].value_counts()[:15]
```

```python
top_15 = df['brand'].value_counts()[:15]
plt.figure(figsize=(12,8))
plt.title("Top 15 brand according to number of vehicles", fontsize=14, fontweight='bold')
top_15.plot.bar()
plt.xlabel('Name of brand',fontsize=12, fontweight='bold')
plt.ylabel('Number of Vehicles',fontsize=12, fontweight='bold')
plt.grid()
# save the plot to file
fig = plt.gcf()
fig.savefig('top-15-regions.png')
plt.show()
```

```python
top_15 = df['brand'].value_counts()[:15]
plt.figure(figsize=(12,8))
plt.title("Top 15 brand according to number of vehicles", fontsize=14, fontweight='bold')
top_15.plot.line()
plt.xlabel('Name of brand',fontsize=12, fontweight='bold')
plt.ylabel('Number of Vehicles',fontsize=12, fontweight='bold')
plt.grid()
# save the plot to file
fig = plt.gcf()
fig.savefig('top-15-regions.png')
plt.show()
```
