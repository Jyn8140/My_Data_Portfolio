```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Set the professional aesthetic
sns.set_theme(style="whitegrid", palette="viridis")

# 2. Distribution of Vehicle Prices (with KDE curve)
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, color='teal', bins=30)
plt.title('Vehicle Market Price Distribution', fontsize=15, pad=20)
plt.xlabel('Price (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

# 3. Correlation Heatmap
# This shows how strongly 'Year' and 'Mileage' relate to 'Price'
plt.figure(figsize=(8, 6))
correlation_matrix = df[['price', 'yom', 'mileage']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Feature Correlation Matrix: Price vs. Age vs. Mileage', fontsize=14)
plt.show()

```
